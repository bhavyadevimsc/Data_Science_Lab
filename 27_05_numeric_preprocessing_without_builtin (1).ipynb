```python
import csv

# Function to check if a value is numeric
def is_numeric(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

# Initialize a dictionary to keep track of null values count for each column
null_counts = {}

# Read the CSV file
with open('numeric.csv', 'r') as file:
    reader = csv.reader(file)
    
    # Get the header
    header = next(reader)
    
    # Initialize the dictionary with column names and zero counts
    for column in header:
        null_counts[column] = 0
    
    # Iterate through each row in the CSV file
    for row in reader:
        for i, value in enumerate(row):
            # Check if the value is non-numeric or empty
            if not is_numeric(value):
                null_counts[header[i]] += 1

# Print the null values count for each column
for column, count in null_counts.items():
    if count>0:
        print(f"{column}: {count} null values")
```

```python
#filling missing vLUES WITH 0
def fill_missing_values(df, fill_value):
    for i in range(len(df)):  # Iterate over rows
        for j in range(len(df.columns)):  # Iterate over columns
            if pd.isnull(df.iloc[i, j]):  # Check if the value is missing
                df.iloc[i, j] = fill_value  # Fill missing value with fill_value

# Sample DataFrame
# df = your_dataframe_here

# Fill missing values with 0
fill_missing_values(df, 0)

# Print DataFrame after filling missing
```

```python
# Define a custom function to drop rows with missing values from a DataFrame
def drop_rows_with_missing_values(df):
    # Initialize an empty list to store rows without missing values
    cleaned_rows = []
    
    # Iterate through each row of the DataFrame
    for i in range(len(df)):
        # Check if the row has any missing values
        has_missing_values = False
        for j in range(len(df.columns)):
            if pd.isnull(df.iloc[i, j]):  # Check if the value is missing
                has_missing_values = True
                break  # Exit loop if a missing value is found
        
        # If the row has no missing values, add it to the cleaned rows list
        if not has_missing_values:
            cleaned_rows.append(df.iloc[i])
    
    # Concatenate the cleaned rows list into a new DataFrame
    cleaned_df = pd.concat(cleaned_rows, axis=1).transpose()
    
    return cleaned_df

# Sample DataFrame
# df = your_dataframe_here

# Drop rows with missing values
cleaned_df = drop_rows_with_missing_values(df)

# Print DataFrame after dropping rows with missing values
print(cleaned_df)
```

```python
def calculate_mean(data):
    total = 0
    count = 0
    for value in data:
        if not pd.isnull(value):  # Exclude missing values from calculation
            total += value
            count += 1
            mean=total / count if count > 0 else 0
    return mean

# Custom function to fill missing values with column means, excluding the last column
def fill_missing_with_means(df):
    for column in df.columns[:-1]:  # Exclude the last column (assuming it contains class labels)
        # Extract non-null values for the current column
        non_null_values = [value for value in df[column] if not pd.isnull(value)]
        # Calculate the mean of non-null values
        column_mean = calculate_mean(non_null_values)
        # Replace missing values in the column with the calculated mean
        for i in range(len(df[column])):
            if pd.isnull(df[column][i]):
                df[column][i] = column_mean
    return df

# Sample DataFrame
# df = your_dataframe_here

# Fill missing values with column means, excluding the last column
df2 = fill_missing_with_means(df)
```

```python
# Custom function to calculate the minimum value of a list
def calculate_min(data):
    min_val = float('inf')  # Initialize with positive infinity
    for value in data:
        if not pd.isnull(value):  # Exclude missing values from calculation
            if value < min_val:
                min_val = value
    return min_val

# Custom function to fill missing values with column minimums, excluding the last column
def fill_missing_with_min(df):
    for column in df.columns[:-1]:  # Exclude the last column (assuming it contains class labels)
        # Extract non-null values for the current column
        non_null_values = [value for value in df[column] if not pd.isnull(value)]
        # Calculate the minimum value of non-null values
        column_min = calculate_min(non_null_values)
        # Replace missing values in the column with the calculated minimum
        for i in range(len(df[column])):
            if pd.isnull(df[column][i]):
                df[column][i] = column_min
    return df

# Sample DataFrame
# df = your_dataframe_here

# Fill missing values with column minimums, excluding the last column
df3 = fill_missing_with_min(df)
```

```python
import numpy as np
import pandas as pd

# Example data
data = pd.read_csv("numeric.csv")

# Separate features and class label
features = data.iloc[:, :-1].apply(pd.to_numeric, errors='coerce')
class_label = data.iloc[:, -1]

# 1. Handling missing values
def handle_missing_values(features):
    # Calculate column-wise means excluding NaN values
    column_means = {}
    for column in features:
        values = [float(value) for value in features[column] if not pd.isna(value)]
        if values:
            column_means[column] = np.mean(values)
        else:
            # If all values in the column are NaN, fill with 0
            column_means[column] = 0

    # Replace NaN values with the mean of the respective column
    for column in features:
        features[column].fillna(column_means[column], inplace=True)
   
    return features

# 2. Remove duplicates
def remove_duplicates(features):
    # Check for duplicates
    if features.duplicated().any():
        print("Duplicates found and will be removed.")
        features = features.drop_duplicates()
    else:
        print("No duplicates found.")
    return features

# Apply handling missing values
features = handle_missing_values(features)
print("Data after handling missing values:")
print(features)

# Apply duplicate removal
features = remove_duplicates(features)
print("Data after removing duplicates:")
print(features)

# Also remove the corresponding rows from the class label
class_label_no_duplicates = class_label.loc[features.index]

# Reattach the class label
final_data = pd.concat([features, class_label_no_duplicates.reset_index(drop=True)], axis=1)
print("Final data with class label:")
print(final_data)
```
